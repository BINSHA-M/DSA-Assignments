# Bargraph of Male VS Female buyers participated in the sales 

```python
Genderdata=data.groupby("Buyer Gender").size().reset_index(name="counts")
Genderdata
```

```python
plt.figure(figsize=(8,8))
Male=Genderdata[["Buyer Gender"]].value_counts()["Male"]
Female=Genderdata[["Buyer Gender"]].value_counts()["Female"]
sns.barplot(x='Buyer Gender',y="counts",data=Genderdata)
plt.title("Male VS Female Buyers participated in sales",fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Counts")
plt.yticks(np.arange(0,6000,500))
plt.grid=(True)

```
